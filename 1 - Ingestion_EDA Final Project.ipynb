### EDA Overview

- id: unique for every restaurant (remove)
- alias: detailed name of restaurant (remove: same information as name)
- name: name of the restaurant (keep)       
- image_url: url to image of restaurant (remove)

- review_count: change data type to numeric (keep: transform)
- categories: create two separate columns, one with an alias and the other with the title (keep: transform)
- rating: average rating of the restaurant (keep)
- coordinates: create two separate columns, one for latitude and one for longitude (keep: transform)
- transactions: empty list (remove)
- price: how expensive the restaurant is (keep: transform)
- location: keep the displayed address only (keep)
- phone: restaurant's phone number (remove)
- display_phone: restaurant's phone number (remove)
- distance: distance (keep)

# Data Pre-Processing

Import the bronze table.

```
data = spark.read.table("df_bronze_1_csv")
data = data.drop('url') # drop the url column
```

```
display(data)
```

```
# Transform the coordinates column to have a separate one for latitude and longitude
import pyspark.sql.functions as f
data = data.withColumn("coordinates", f.regexp_replace('coordinates',r'[{}"":longitudelatitude]',""))
data = data.withColumn("coordinates2", data["coordinates"])

split_col = f.split(data['coordinates'], ',')
data = data.withColumn("coordinates", split_col.getItem(0))
data = data.withColumnRenamed("coordinates","latitude")

split_col = f.split(data['coordinates2'], ',')
data = data.withColumn("coordinates2", split_col.getItem(1))
data = data.withColumnRenamed("coordinates2","longitude")
```

```
# Extract the postal code fom the location that we will use later
data = data.withColumn("postal_code", data["location"])
data = data.withColumn("postal_code", f.regexp_replace('postal_code',r'[_":{}abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ]',""))

split_col = f.split(data['postal_code'], ',')
data = data.withColumn("postal_code", split_col.getItem(0))

data = data.drop('location')
```

```
# Get the restaurant categories
pandas_d = data.toPandas()
pandas_d.iloc[0:50,2] = "Spanish"
pandas_d.iloc[50:100,2] = "Burgers"
pandas_d.iloc[100:150,2] = "Italian"
pandas_d.iloc[150:200,2] = "Mexican"

data = spark.createDataFrame(pandas_d)
```

```
display(data)
```

```
# Let us  check the data types
data.dtypes
```

```
# We will put the numerical columns in the right type
data = data.withColumn("rating", data["rating"].cast('float'))
data = data.withColumn("review_count", data["review_count"].cast('float'))
data = data.withColumn("price", data["price"].cast('int'))
data = data.withColumn("latitude", data["latitude"].cast('float'))
data = data.withColumn("longitude", data["longitude"].cast('float'))
data = data.withColumn("postal_code", data["postal_code"].cast('int'))
```
